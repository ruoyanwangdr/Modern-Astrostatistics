## MODELLING AND INFERRING STELLAR PROPERTIES
### For T0 = 10, σ = 1, A = 1, plot the distribution function of stellar luminosities.

```python
def prob(L):
    T0=10
    sigma=1
    A=1
    T=(L/A)**(1/4)
    P_L=np.exp(-((T-T0)**2/2*sigma**2))*L**(-3/4)*(1/(np.sqrt(2*np.pi*sigma**2)))*(1/(4*A))
    return P_L
```

```python
L=np.linspace(500,30000,100)
P_L=prob(L)
plt.plot(L,P_L,'.')
plt.xlabel('luminosity', size=20)
plt.ylabel('P(L)', size=20)
```

### Generate random samples from P(L), histogram them, and show that the histogram approximates P(L) for sufficiently many samples.

```python
from random import uniform
n=10000
h=np.zeros(n)
for i in range(n):
    L=uniform(500,30000)
    h[i]=prob(L)
bins=plt.hist(h, 100)
```

## Satellities: the next generation


```python
tau,mu,sigma=1,10,1
theta=np.random.normal(mu, tau, 1000)
y = np.random.normal(theta, sigma, 1000)
bins = plt.hist(theta, 100, density=True, alpha=0.5, label='prior')
bins = plt.hist(y*theta, 100, density=True, alpha=0.5, label='posterior')
bins = plt.hist(y, 100, density=True, alpha=0.5, label='data')
#plt.xlim(-10,10)
plt.legend()
```
